# PI-DeepONet-RAR — Kaggle Runner
**Workflow:** Repo mounted as dataset → run experiment → push outputs to GitHub

Before running:
1. Add your repo as a Kaggle dataset (Upload → New Dataset → link GitHub repo)
2. Add `GITHUB_TOKEN` to Kaggle Secrets (Add-ons → Secrets)
3. Set `GITHUB_REPO` below to `your-username/PI-DeepONet-RAR`

```python
# ── Config ───────────────────────────────────────────────────────────────────
GITHUB_REPO   = "your-username/PI-DeepONet-RAR"   # <-- change this to your repo
DATASET_PATH  = "/kaggle/input/pi-deeponet-rar"   # Kaggle mounts dataset here
OUTPUT_DIR    = "/kaggle/working"
```

```python
# ── Mount repo into Python path ───────────────────────────────────────────────
import sys, os, glob

if os.path.exists(DATASET_PATH):
    sys.path.insert(0, DATASET_PATH)
    print(f"Repo path added: {DATASET_PATH}")
else:
    matches = glob.glob("/kaggle/input/*pi-deeponet*")
    if matches:
        DATASET_PATH = matches[0]
        sys.path.insert(0, DATASET_PATH)
        print(f"Detected dataset path: {DATASET_PATH}")
    else:
        print(f"⚠️ Dataset path {DATASET_PATH} not found!")
```

```python
# ── Install any missing deps ──────────────────────────────────────────────────
!pip install pyyaml scipy torch matplotlib pandas -q
```

```python
# ── Run experiment ────────────────────────────────────────────────────────────
import subprocess

# Run experiment (remove '--quick' for full experiment suite run)
cmd = [sys.executable, os.path.join(DATASET_PATH, "kaggle_run_all.py"), "--quick"]

print("Starting experiment...")
process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True, bufsize=1, cwd=OUTPUT_DIR)
for line in process.stdout:
    print(line, end="")
process.wait()
print(f"\nExperiment completed with exit code: {process.returncode}")
```

```python
# ── Push outputs to GitHub ────────────────────────────────────────────────────
from kaggle_secrets import UserSecretsClient
from utils.io import push_to_github

token = UserSecretsClient().get_secret("GITHUB_TOKEN")

push_to_github(
    token=token,
    repo=GITHUB_REPO,
    source_dirs=[
        f"{OUTPUT_DIR}/results",
    ],
    commit_message="auto: PI-DeepONet-RAR experiment outputs from Kaggle",
)
```
